```python

non_string_count = df['Quantity'].apply(lambda x: not isinstance(x, str)).sum()
non_string_count
```

```python
def convert_quantity(row):
    try:
        return int(row.split()[0])
    except:
        return None 
```

```python
df['Quantity'] = df['Quantity'].apply(convert_quantity)

# Drop rows with invalid (None) Quantity
df = df.dropna(subset=['Quantity'])
df['Quantity'] = df['Quantity'].astype(int)
df
```
